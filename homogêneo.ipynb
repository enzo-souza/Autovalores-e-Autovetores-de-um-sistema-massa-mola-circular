# Introdução

O problema massa mola é um objeto de estudo do comportamento de um sistema que possui molas ligadas à molas. Com ele, é possível quantificar as frequências possíveis e todas as possibilidades de movimento para cada frequência em cada massa.

Nesta questão, devemos estudar um sistema em que a última mola está ligada à primeira, ou seja, um sistema circular:

![alt text](image-2.png)


# Metodologia

Utilizamos as bibliotecas Numpy para realizar os calculos de algebra liner e a biblioteca matplotlib para plotar os gráficos.

As matrizes são criadas via código digitado.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define os valores de  massa e da constante elástica:

massa = 1
k = 1
n = int(input())


#constroi tres matrizes homogêneas que serão usadas para os calculos da matriz a ser diagonalizada.

# Cria uma lista de números de 0 a n-1
a = []
for z in range(0,n):
    a.append(z)
print(a)
print(len(a))

matriz_massa =[]
matriz_k = []
matriz = []
for z in range (0,n):
    matriz_massa.append([])
    matriz_k.append([])
    matriz.append([])
    for y in range(0,n):
        matriz_massa[z].append(massa)
        matriz_k[z].append(k)
        matriz[z].append(0)

#printa as matrizes

'''print(matriz_massa)
print(matriz_k)
print(matriz)'''
```

As equações dinâmicas do sistema são encontradas seguindo a funçãos dadas pelo professor. Porém, para o nosso caso em que a última massa está ligada à primeira, foi necessário utilizar a artimanha do operador "%" que encontra o valor equivalente em um dado conjunto. Exemplo: Se meu conjunto A é dado por [0,4], a posição A[5] == 0

![alt text](image-3.png)


Neste primeiro momento, os cálculos são realizados para um sistema homogênio, ou seja, todas as massas e constantes elásticas são iguais:

```python
#calcula os autovalores para um n definido na celula anterior em um 
# sistema homogeneo em que o a ultima massa está ligada a primeira.

a = []
for z in range(0,n):
    a.append(z)
#print(a)


for i in range(0,len(a)):
    for j in range (0,len(a)):
        
        #Elenca os valores da diagonal pr
        if i==(j%len(a)):
            matriz[i][j] = ((matriz_k[i][(j-1)%len(a)] + matriz_k[i][(j+1)%len(a)]) / matriz_massa[i][j])
        
        #Elenca os valores da diagonal secundária superior
        if i==((j-1)%len(a)):
            matriz[i][j] = -1*(matriz_k[i][j]) / (matriz_massa[i][j])

        #Elenca os valores da diagonal secundária inferior
        if i==((j+1)%len(a)):
            matriz[i][j] = -1*(matriz_k[i][j]) / (matriz_massa[i][j])

# Imprime a matriz resultante
print()
for p in matriz:
    print(p)
```

Com as matrizes prontas, basta utilizas a função eig para diagonalizá-las e encontrar os autovalores (frequências possíveis) e autovetores (modo do conjunto):

```python
#Encontra os autovalores e autovetores da matriz
import matplotlib.pyplot as plt

# Calcula os autovalores da matriz usando a biblioteca NumPy eig:
autovalores, autovetores  = np.linalg.eig(matriz)

#Na biblioteca eig, os autovetores são retornados na coluna, então precisamos transpor a matriz de autovetores para obter os autovetores como linhas.
autovetores.T

print("autovalores:", autovalores)
print("autovetores:","\n", autovetores)

#Ordena os autovalores e autovetores
idx = autovalores.argsort()
print('\n',idx)
print(autovetores[:,idx])
autovetores = np.round(autovetores, decimals=5)
```

# Resultados

# HISTOGRAMA

```python
# Plota a densidade de estados (histograma dos autovalores)
plt.figure(figsize=(6,4))
plt.hist(autovalores, bins=9, density=True, color='blue')
plt.xlabel('Autovalores')
plt.ylabel('Densidade de estados')
plt.title('Densidade de estados dos autovalores')
plt.grid(True)
plt.show()
```

# 5 PRIMEIROS AUTOVALORES

```python
# Plota o primeiro autovetor em função de uma reta real x
x = np.arange(n)
#plt.figure(figsize=(20,8))

for i in range(5):
    plt.plot(x, autovetores[:,idx[i]], marker='o', alpha=0.7, linestyle='-', color='blue')
    plt.xlabel('x (reta real)')
    plt.ylabel('Autovetor')
    plt.title(f'Frequência {autovalores[idx[i]]:.3f}')
    plt.grid(True)
    plt.show()
```
